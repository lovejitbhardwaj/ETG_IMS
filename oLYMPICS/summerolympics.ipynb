# Summer Olympics Data Analysis Assignment

### 1. In how many cities Summer Olympics is held so far?

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv("summer.csv")
```

```python
q1 = []
for city in df['City'].unique():
    q1.append(city)
len(q1)
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
data = df[['Sport', 'Medal']]
data = data[data['Medal'] ==  'Gold']
q2 = []
for sport in data['Sport'].unique():
    q2.append([sport , len(data[data['Sport']  == sport])])

q2 = pd.DataFrame(q2,columns = ['Sport','Gold Medal']).sort_values(by='Gold Medal', ascending=False)
q2.head().plot(x = 'Sport', y = 'Gold Medal', kind = 'bar',color=['gold'])
```

### 3. Which sport is having most number of medals so far? (Top 5)

```python
data = df[['Sport', 'Medal']]
q3 = []
for sport in data['Sport'].unique():
    q3.append([sport , len(data[data['Sport']  == sport])])

q3 = pd.DataFrame(q3,columns = ['Sport','No of Medals']).sort_values(by='No of Medals', ascending=False)
q3.head().plot(x = 'Sport', y = 'No of Medals', kind = 'bar')
```

### 4. Which player has won most number of medals? (Top 5)

```python
data = df[['Athlete', 'Medal']]
q4 = []
for athlete in data['Athlete'].unique():
    q4.append([athlete , len(data[data['Athlete']  == athlete])])

q4 = pd.DataFrame(q4,columns = ['Athlete','No of Medals']).sort_values(by='No of Medals', ascending=False)
q4.head().plot(x = 'Athlete', y = 'No of Medals', kind = 'bar',color=(0.1, 0.1, 0.1, 0.1),  edgecolor='blue')
```

### 5. Which player has won most number Gold Medals of medals? (Top 5)

```python
data = df[['Athlete', 'Medal']]
data = data[data['Medal'] ==  'Gold']
q5 = []
for athlete in data['Athlete'].unique():
    q5.append([athlete , len(data[data['Athlete']  == athlete])])

q5 = pd.DataFrame(q5,columns = ['Athlete','Gold Medals']).sort_values(by='Gold Medals', ascending=False)
q5.head().plot(x = 'Athlete', y = 'Gold Medals', kind = 'bar')
```

### 6. In which year India won first Gold Medal in Summer Olympics?

```python
data = df[['Year', 'Country', 'Medal']]
data = data[data['Medal'] ==  'Gold']
data = data[data['Country'] ==  'IND']

q6 = pd.DataFrame(data,columns = ['Year','Country','Medal']).sort_values(by='Year')
print(q6.iloc[0][0])
```

### 7. Which event is most popular in terms on number of players? (Top 5)

```python
data = df[['Event', 'Athlete']]
q7 = []
for event in data['Event'].unique():
    q7.append([event , len(data[data['Event']  == event])])

q7 = pd.DataFrame(q7,columns = ['Event','Players']).sort_values(by='Players', ascending=False)
q7.head().plot(x = 'Event', y = 'Players', kind = 'bar')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
data = df[['Sport', 'Gender', 'Medal']]
data = data[data['Gender'] ==  'Women']
data = data[data['Medal'] ==  'Gold']
q8 = []
for sport in data['Sport'].unique():
    q8.append([sport, len(data[data['Sport']  == sport])])

q8 = pd.DataFrame(q8,columns = ['Sport','No of Medals']).sort_values(by='No of Medals', ascending=False)
q8.head().plot(x = 'Sport', y = 'No of Medals', kind = 'bar')
```
